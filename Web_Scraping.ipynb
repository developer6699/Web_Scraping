```python
# Collecting all the urls from onecle
links0 = []
for link in soup.find_all('a', href=True):
    if link['href'].startswith("https:"):
        #print(link['href'])
        links0.append(link['href'])
    elif link['href'].startswith("http:"):
        #print(link['href'])
        links0.append(link['href'])
    elif link['href'].startswith("//"):
        #print("https:"+link1['href'])
        links0.append("https:"+link['href'])
    else:
        #print("https://"+link['href'])
        links0.append("https://"+link['href'])
```

```python
links = list(set(links0))
links
```

```python
def append_multiple_lines(file_name, lines_to_append):
    with io.open(file_name, 'w+', encoding="utf-8") as f:
  
        # write elements of list
        for items in lines_to_append:
            f.write('%s\n' %items)

        #print("File appended successfully")

    # close the file
    f.close()
    
    #import os
    #os.rename('file.txt',file_name)
```

```python
def append_multiple_lines1(file_name, lines_to_append):
    # Open the file in append & read mode ('a+')
    with io.open(file_name, "w+", encoding="utf-8") as file_object:
        appendEOL = False
        # Move read cursor to the start of file.
        file_object.seek(0)
        # Check if file is not empty
        data = file_object.read(100)
        if len(data) > 0:
            appendEOL = True
        # Iterate over each string in the list
        for line in lines_to_append:
            # If file is not empty then append '\n' before first line for
            # other lines always append '\n' before appending line
            if appendEOL == True:
                file_object.write("\n")
            else:
                appendEOL = True
            # Append element at the end of file
            file_object.write(line)
```
